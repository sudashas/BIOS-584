```python
#Research question 1: How has the trade volume of staple agricultural products (wheat, maize, barley) changed between the United Kingdom and the European Union post-Brexit?

#summary statistics: counts 
print(brexit_trade[["refYear","cmdDesc","partnerDesc","netWgt"]].describe())
print(brexit_trade["cmdDesc"].value_counts())
print(brexit_trade["partnerDesc"].value_counts())
print(brexit_trade["refYear"].value_counts().sort_index())
commodities = ["Wheat and meslin", "Barley", "Maize (corn)"]
#Summary statistics: mean export by year
filtered = brexit_trade[brexit_trade["cmdDesc"].isin(commodities)]
filtered["refYear"] = pd.to_numeric(filtered["refYear"], errors="coerce")
filtered["netWgt"] = pd.to_numeric(filtered["netWgt"], errors="coerce")
mean_by_year = filtered.groupby(["refYear", "cmdDesc"])["netWgt"].mean().reset_index()
pivot_means = mean_by_year.pivot(index="refYear", columns="cmdDesc", values="netWgt")
print(pivot_means)




#Plot: based on professor advice, multi-line graph with year on x-axis and marking when brexit was
brexit_trade["netWgt"] = pd.to_numeric(brexit_trade["netWgt"], errors="coerce")

# Filter for the staple crops to make sure there are no extra lines 
commodities = ["Wheat and meslin", "Barley", "Maize (corn)"]
filtered = brexit_trade[brexit_trade["cmdDesc"].isin(commodities)]

# Group by year and commodity
grouped = filtered.groupby(["refYear", "cmdDesc"])["netWgt"].sum().reset_index()
pivot_df = grouped.pivot(index="refYear", columns="cmdDesc", values="netWgt")

# Plot
plt.style.use("seaborn-v0_8")
fig, ax = plt.subplots(figsize=(10, 6))

for commodity in commodities:
    if commodity in pivot_df.columns:
        ax.plot(pivot_df.index, pivot_df[commodity], label=commodity)

ax.axvline(x=2020, color="red", linestyle="--", label="UK leaves EUSMC")
ax.axvline(x=2016, color="orange", linestyle="--", label="Brexit vote")
ax.set_title("UK Exports of Staple Agricultural Products to EU (by Year)")
ax.set_xlabel("Year")
ax.set_ylabel("Net Weight (kg)")
ax.legend()
ax.grid(True)
plt.tight_layout()
plt.savefig("RQ1_agricultural_exports.png", dpi=300) 
plt.show()
```

```python
#Research question 2 

#summary statistics 
#Missing values
missing_counts = sahel_trade[["refYear", "reporterDesc", "cmdDesc", "netWgt"]].isnull().sum()
print(missing_counts)

#Overall average net weight per crop
overall_avg = sahel_trade.groupby("cmdDesc")["netWgt"].mean().reset_index()
print(overall_avg)

#Yearly average net weight per crop
yearly_avg = sahel_trade.groupby(["refYear", "cmdDesc"])["netWgt"].mean().reset_index()
print(yearly_avg)


#plot
import matplotlib.pyplot as plt

sahel_trade["refYear"] = pd.to_numeric(sahel_trade["refYear"], errors="coerce")
sahel_trade["netWgt"] = pd.to_numeric(sahel_trade["netWgt"], errors="coerce")
#used AI help to make this function, I tried using the internet guide of pyplot to figure out how to make pie charts
def get_crop_group(desc):
    if "Ground-nuts" in desc:
        return "Groundnuts"
    elif "Grain sorghum" in desc:
        return "Sorghum"
    elif "Buckwheat, millet" in desc:
        return "Millet"
    elif "sesamum seeds" in desc:
        return "Sesame"
    elif "oleaginous fruits" in desc:
        return "Other oilseeds"
    else:
        return None

sahel_trade["crop"] = sahel_trade["cmdDesc"].apply(get_crop_group)
sahel_trade = sahel_trade[sahel_trade["crop"].notnull()]

windows = {
    "20082011": (2008, 2011),
    "2012": (2012, 2012),
    "20132019": (2013, 2019)
}

fig, axes = plt.subplots(1, 3, figsize=(18, 6))
plt.style.use("seaborn-v0_8")

for ax, (label, (start, end)) in zip(axes, windows.items()):
    subset = sahel_trade[(sahel_trade["refYear"] >= start) & (sahel_trade["refYear"] <= end)]
    totals = subset.groupby("crop")["netWgt"].sum()
    ax.pie(totals, labels=totals.index, autopct="%1.1f%%", startangle=90)
    ax.set_title(f"Crop Export Share ({label})")

plt.tight_layout()
plt.savefig("RQ2_NativeCropExports.png", dpi=300)
plt.show()
```
