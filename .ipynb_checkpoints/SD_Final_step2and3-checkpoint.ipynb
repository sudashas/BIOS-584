```python
#Research Question 3: How did the trade war between the United States and China between 2018 and 2020 alter Chinas import patterns of Soybeans and other American agricultural products? 
# RQ3: aggregate into time windows and plot (fixed)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Ensure numeric year/value
china_trade["refYear"] = pd.to_numeric(china_trade["refYear"], errors="coerce")
china_trade["primaryValue"] = pd.to_numeric(china_trade["primaryValue"], errors="coerce")

# Classify commodity (case-insensitive)
def classify_commodity(desc):
    if pd.isna(desc):
        return None
    d = desc.lower()
    if "soya" in d or "soy" in d:
        return "Soybeans"
    if "maize" in d or "corn" in d:
        return "Corn"
    if "meat of swine" in d or "pork" in d:
        return "Pork"
    if "cotton" in d:
        return "Cotton"
    return None

china_trade["commodity"] = china_trade["cmdDesc"].apply(classify_commodity)
china_trade = china_trade[china_trade["commodity"].notnull()].copy()

# Define periods
def year_to_period(y):
    if 2015 <= y <= 2017:
        return "20152017"
    if 2018 <= y <= 2020:
        return "20182020"
    if 2021 <= y <= 2024:
        return "20212024"
    return np.nan

china_trade["period"] = china_trade["refYear"].apply(year_to_period)
china_trade = china_trade[china_trade["period"].notnull()]

# Aggregate total trade value per commodity  period
agg = china_trade.groupby(["commodity", "period"])["primaryValue"].sum().reset_index()

# Pivot to wide format for plotting
plot_df = agg.pivot(index="commodity", columns="period", values="primaryValue").fillna(0)

# Optional: reorder columns to desired time order
plot_df = plot_df[["20152017", "20182020", "20212024"]]

# Rename columns for nicer legend/labels
custom_labels = {
    "20152017": "20152017 (before tariffs)",
    "20182020": "20182020 (trade war)",
    "20212024": "20212024 (after tariffs)"
}
plot_df.columns = [custom_labels[c] for c in plot_df.columns]

# Plot
plt.style.use("seaborn-v0_8")
fig, ax = plt.subplots(figsize=(10, 6))
plot_df.T.plot(kind="bar", ax=ax)   # transpose so periods are x-axis
ax.set_title("US Exports to China by Commodity and Time Window (Trade Value)")
ax.set_ylabel("Total Trade Value (USD)")
ax.set_xlabel("Time Window")
ax.legend(title="Commodity", bbox_to_anchor=(1.02, 1), loc="upper left")
plt.tight_layout()
plt.savefig("RQ3_TradeWar.png", dpi=300)
plt.show()

# Print the aggregated table for inspection
print("\nAggregated trade values (USD):")
print(plot_df)
```

```python
#Research Question 1, step 3 
# Wilcoxon signed-rank test per commodity (paired by partner country)
import pandas as pd
import numpy as np
from scipy.stats import wilcoxon

brexit_trade["refYear"] = pd.to_numeric(brexit_trade["refYear"], errors="coerce")
brexit_trade["netWgt"]  = pd.to_numeric(brexit_trade["netWgt"], errors="coerce")

commodities = ["Wheat and meslin", "Barley", "Maize (corn)"]

results = []
for comm in commodities:
    sub = brexit_trade[brexit_trade["cmdDesc"] == comm].copy()
    before = sub[sub["refYear"] < 2020].groupby("partnerDesc")["netWgt"].mean()
    after  = sub[sub["refYear"] >= 2020].groupby("partnerDesc")["netWgt"].mean()
    common = before.index.intersection(after.index)
    paired = pd.DataFrame({
        "before": before.loc[common],
        "after":  after.loc[common]
    }).dropna()

    n_pairs = paired.shape[0]
    if n_pairs < 3:
        results.append({
            "commodity": comm,
            "n_pairs": n_pairs,
            "wilcoxon_stat": np.nan,
            "p_value": np.nan,
            "median_before": paired["before"].median() if n_pairs>0 else np.nan,
            "median_after": paired["after"].median() if n_pairs>0 else np.nan,
            "note": "Not enough paired countries (need >=3)"
        })
        continue

    
    diffs = paired["after"] - paired["before"]
    if np.allclose(diffs, 0, equal_nan=True):
        results.append({
            "commodity": comm,
            "n_pairs": n_pairs,
            "wilcoxon_stat": 0.0,
            "p_value": 1.0,
            "median_before": paired["before"].median(),
            "median_after": paired["after"].median(),
            "note": "All paired differences are zero"
        })
        continue

    try:
        stat, p = wilcoxon(paired["before"], paired["after"], zero_method="wilcox", alternative="two-sided")
    except Exception as e:
        stat, p = wilcoxon(paired["before"], paired["after"])

    med_before = paired["before"].median()
    med_after  = paired["after"].median()
    pct_change = (med_after - med_before) / med_before * 100 if (med_before not in (0, np.nan)) else np.nan

    #labels
    results.append({
        "commodity": comm,
        "n_pairs": n_pairs,
        "wilcoxon_stat": stat,
        "p_value": p,
        "median_before": med_before,
        "median_after": med_after,
        "median_pct_change": pct_change,
        "note": ""
    })

#printing
results_df = pd.DataFrame(results).set_index("commodity")
pd.options.display.float_format = "{:,.4f}".format
print(results_df)

#Interpretation 
#When running the Wilcoxon signed rank test, our null hypothesis is that the difference between the export values from the UK for each staple crop is zero before and after brexit
#based on our results, only wheat was statistically significant at the 95% interval, with a percent reduction of 82% from before brexit and after brexit, p=0.0186 
#although not statistically significiant, barley saw a 37 percent reduction before and after brexit and surprisingly corn saw a 590 percent increase
```

```python
#Research Question 2 Step 3 
# Wilcoxon signed-rank test per crop (paired by reporter country)
import pandas as pd
import numpy as np
from scipy.stats import wilcoxon

sahel_trade["refYear"] = pd.to_numeric(sahel_trade["refYear"], errors="coerce")
sahel_trade["netWgt"]  = pd.to_numeric(sahel_trade["netWgt"], errors="coerce")

crops = sahel_trade["crop"].unique().tolist() #lots of repeated values, needs to be unique 

results = []
pre_mask  = (sahel_trade["refYear"] >= 2008) & (sahel_trade["refYear"] <= 2011)
post_mask = (sahel_trade["refYear"] >= 2013) & (sahel_trade["refYear"] <= 2019)

for crop in sorted(crops):
    sub = sahel_trade[sahel_trade["crop"] == crop].copy()
    before = sub[pre_mask].groupby("reporterDesc")["netWgt"].mean()
    after  = sub[post_mask].groupby("reporterDesc")["netWgt"].mean()
    common = before.index.intersection(after.index)
    paired = pd.DataFrame({"before": before.loc[common], "after": after.loc[common]}).dropna()

    n_pairs = paired.shape[0]
    if n_pairs < 3:
        results.append({
            "crop": crop,
            "n_pairs": n_pairs,
            "wilcoxon_stat": np.nan,
            "p_value": np.nan,
            "median_before": paired["before"].median() if n_pairs>0 else np.nan,
            "median_after": paired["after"].median() if n_pairs>0 else np.nan,
            "note": "Too few paired countries (need >=3)"
        })
        continue

    diffs = paired["after"] - paired["before"]
    if np.allclose(diffs, 0, equal_nan=True):
        results.append({
            "crop": crop,
            "n_pairs": n_pairs,
            "wilcoxon_stat": 0.0,
            "p_value": 1.0,
            "median_before": paired["before"].median(),
            "median_after": paired["after"].median(),
            "note": "All paired differences are zero"
        })
        continue

    try:
        stat, p = wilcoxon(paired["before"], paired["after"], zero_method="wilcox", alternative="two-sided")
    except Exception:
        stat, p = wilcoxon(paired["before"], paired["after"])

    med_before = paired["before"].median()
    med_after  = paired["after"].median()
    pct_change = (med_after - med_before) / med_before * 100 if (med_before not in (0, np.nan)) else np.nan

    results.append({
        "crop": crop,
        "n_pairs": n_pairs,
        "wilcoxon_stat": stat,
        "p_value": p,
        "median_before": med_before,
        "median_after": med_after,
        "median_pct_change": pct_change,
        "note": ""
    })

results_df = pd.DataFrame(results).set_index("crop")
pd.options.display.float_format = "{:,.4f}".format
print(results_df)


#NEW SPAGHETTI AND PAIRED DIFFERENCE PLOTS 
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

for crop in sorted(sahel_trade["crop"].unique()):
    sub = sahel_trade[sahel_trade["crop"] == crop].copy()
    if sub.empty:
        continue

    # Spaghett
    plt.figure(figsize=(10,6))
    sns.lineplot(data=sub, x="refYear", y="netWgt", hue="reporterDesc", estimator=None, alpha=0.6, legend=False)
    mean_series = sub.groupby("refYear")["netWgt"].mean().reset_index()
    sns.lineplot(data=mean_series, x="refYear", y="netWgt", color="black", linewidth=2.5, label="Mean across countries")
    plt.title(f"{crop}  Yearly trade weight by country (spaghetti)")
    plt.xlabel("Year")
    plt.ylabel("Net weight")
    plt.axvline(2012, color="red", linestyle="--", label="2012 crisis")
    plt.legend(bbox_to_anchor=(1.02,1), loc="upper left")
    plt.tight_layout()
    plt.show()

    # Paired-difference plot
    pre = sub[(sub["refYear"] >= 2008) & (sub["refYear"] <= 2011)].groupby("reporterDesc")["netWgt"].mean()
    post = sub[(sub["refYear"] >= 2013) & (sub["refYear"] <= 2019)].groupby("reporterDesc")["netWgt"].mean()
    common = pre.index.intersection(post.index)
    paired = pd.DataFrame({"before": pre.loc[common], "after": post.loc[common]}).dropna()
    if paired.shape[0] == 0:
        continue

    paired["diff"] = paired["after"] - paired["before"]
    paired_sorted = paired.sort_values("diff")

    plt.figure(figsize=(8, max(4, paired_sorted.shape[0]*0.3)))
    plt.hlines(y=paired_sorted.index, xmin=paired_sorted["before"], xmax=paired_sorted["after"], color="gray", alpha=0.7)
    plt.scatter(paired_sorted["before"], paired_sorted.index, color="blue", label="Before (2008-2011)")
    plt.scatter(paired_sorted["after"], paired_sorted.index, color="orange", label="After (2013-2019)")
    plt.axvline(0, color="black", linewidth=0.8)
    plt.title(f"{crop}  Paired country means: before vs after")
    plt.xlabel("Mean net weight")
    plt.legend()
    plt.tight_layout()
    plt.show()

#interpretation 
#Per the WSR test, the two crops that saw a statistically significant increase from during the 2012 during and after was groundnuts with an 87 percent increase and sesane with a 119 percent increase 
```

```python
#Research question 3, step 3 
# Normalize plot_df column names and run the repeated-measures analyses
import numpy as np
import pandas as pd

#change the column labels back from the plots done in step 2
col_map = {
    '20152017 (before tariffs)': '2015_2017',
    '20182020 (trade war)':     '2018_2020',
    '20212024 (after tariffs)':  '2021_2024'
}

plot_df = plot_df.rename(columns=col_map)


canonical = ['2015_2017','2018_2020','2021_2024']
for c in canonical:
    if c not in plot_df.columns:
        plot_df[c] = np.nan
plot_df = plot_df[canonical]  # reorder

#check to see if columns back to normal because they weren't earlier
print("Normalized plot_df columns:", list(plot_df.columns))
print("\nAggregated table (commodity  normalized period):\n")
print(plot_df)

#building models for linear regression
rm_df = plot_df.reset_index().melt(id_vars='commodity', var_name='period', value_name='value')
# readable labels for plots if needed
period_label_map = {'2015_2017':'20152017','2018_2020':'20182020','2021_2024':'20212024'}
rm_df['period_label'] = rm_df['period'].map(period_label_map)
rm_df = rm_df.sort_values(['commodity','period']).reset_index(drop=True)


from scipy.stats import friedmanchisquare, wilcoxon
from statsmodels.stats.anova import AnovaRM
import statsmodels.formula.api as smf


a = plot_df['2015_2017'].values
b = plot_df['2018_2020'].values
c = plot_df['2021_2024'].values
mask = ~np.isnan(a) & ~np.isnan(b) & ~np.isnan(c)
if mask.sum() >= 3:
    stat_f, p_f = friedmanchisquare(a[mask], b[mask], c[mask])
    print(f"\nFriedman test (n={mask.sum()}): chi-square={stat_f:.4f}, p={p_f:.4f}")
else:
    print("\nFriedman test: not enough complete commodities (need >=3 complete rows).")

# Repeated-measures ANOVA (AnovaRM), used ai help 
complete_counts = rm_df.groupby("commodity")["period"].nunique()
if complete_counts.min() == 3:
    try:
        aov = AnovaRM(rm_df, depvar="value", subject="commodity", within=["period"]).fit()
        print("\nRepeated-measures ANOVA (AnovaRM) results:\n")
        print(aov)
    except Exception as e:
        print("\nAnovaRM failed:", e)
else:
    print("\nAnovaRM skipped: data are unbalanced (not every commodity has all 3 periods).")

rm_df['period'] = pd.Categorical(rm_df['period'], categories=canonical, ordered=True)
try:
    lmm = smf.mixedlm("value ~ C(period)", rm_df, groups=rm_df["commodity"])
    lmm_res = lmm.fit(reml=False)
    print("\nLinear Mixed Model (random intercept for commodity):\n")
    print(lmm_res.summary())
except Exception as e:
    print("\nLMM failed:", e)

try:
    ols = smf.ols("value ~ C(period) + C(commodity)", data=rm_df).fit()
    print("\nOLS with commodity fixed effects (period coefficients relative to 2015_2017):\n")
    print(ols.summary())
except Exception as e:
    print("\nOLS failed:", e)

# Pairwise Wilcoxon post-hoc with Bonferroni correction, needed ai help 
pairs = [('2015_2017','2018_2020'), ('2015_2017','2021_2024'), ('2018_2020','2021_2024')]
pair_results = []
for a_name, b_name in pairs:
    if a_name in plot_df.columns and b_name in plot_df.columns:
        x = plot_df[a_name].values
        y = plot_df[b_name].values
        valid = ~np.isnan(x) & ~np.isnan(y)
        if valid.sum() >= 3:
            stat_w, p_w = wilcoxon(x[valid], y[valid], alternative="two-sided")
            p_bonf = min(p_w * len(pairs), 1.0)
        else:
            stat_w, p_w, p_bonf = np.nan, np.nan, np.nan
    else:
        stat_w, p_w, p_bonf = np.nan, np.nan, np.nan
    pair_results.append({
        "pair": f"{a_name} vs {b_name}",
        "n_pairs": int(valid.sum()) if 'valid' in locals() else 0,
        "wilcoxon_stat": stat_w,
        "p_value": p_w,
        "p_bonf": p_bonf
    })
pair_df = pd.DataFrame(pair_results)
print("\nPairwise Wilcoxon results (Bonferroni corrected p):\n")
print(pair_df)

period_summary = rm_df.groupby("period_label")["value"].agg(["mean","median","std","count"]).reindex(["20152017","20182020","20212024"])
print("\nPeriod summary (across commodities):\n")
print(period_summary)

#interp 
#The linear mixed model shows a significant positive effect for 20212024 versus 20152017 (coef = 10,308,044,940; p = 0.034). In the OLS with commodity fixed effects, Soybeans have a significantly higher trade value than the reference (coef = 36,030,000,000; p = 0.001).
```

```python
#Research Question 4, step 3
# Paired Wilcoxon and paired t-test
import numpy as np
import pandas as pd
from scipy.stats import wilcoxon, ttest_rel, shapiro, t

#adjusting columns and dataframe
df = ukraine_trade.copy()
df['refYear'] = pd.to_numeric(df['refYear'], errors='coerce')
df['netWgt']  = pd.to_numeric(df['netWgt'], errors='coerce')

#defining my two time periods of pre and post war 
pre_mask  = df['refYear'].between(2015, 2021)
post_mask = df['refYear'].between(2022, 2024)

#getting the mean by country
pre_means  = df.loc[pre_mask].groupby('partnerDesc')['netWgt'].mean()
post_means = df.loc[post_mask].groupby('partnerDesc')['netWgt'].mean()

#pairing
common = pre_means.index.intersection(post_means.index)
paired = pd.DataFrame({'pre': pre_means.loc[common], 'post': post_means.loc[common]}).dropna()

n_pairs = paired.shape[0]
print(f"Paired countries included: {n_pairs}\n")
if n_pairs == 0:
    raise ValueError("No countries have both pre and post data. Cannot run paired tests.")

paired['diff'] = paired['post'] - paired['pre']
mean_diff = paired['diff'].mean()
median_pre = paired['pre'].median()
median_post = paired['post'].median()
median_diff = paired['diff'].median()

# Shapiro test for normality of differences (only meaningful if n >= 3)
if n_pairs >= 3:
    sh_stat, sh_p = shapiro(paired['diff'])
    print(f"Shapiro-Wilk test for normality of differences: W={sh_stat:.4f}, p={sh_p:.4f}")
else:
    sh_stat, sh_p = np.nan, np.nan
    print("Shapiro-Wilk test skipped (need >=3 pairs).")

# Wilcoxon signed-rank test
if n_pairs >= 3:
    w_stat, w_p = wilcoxon(paired['pre'], paired['post'], alternative='two-sided')
else:
    w_stat, w_p = np.nan, np.nan

# Paired t-test (parametric on sharpiro walk test)
if n_pairs >= 2:
    t_stat, t_p = ttest_rel(paired['pre'], paired['post'], nan_policy='omit')
else:
    t_stat, t_p = np.nan, np.nan

# 95% CI for mean difference (paired t CI)
if n_pairs >= 2:
    se = paired['diff'].std(ddof=1) / np.sqrt(n_pairs)
    dfree = n_pairs - 1
    t_crit = t.ppf(0.975, dfree)
    ci_lower = mean_diff - t_crit * se
    ci_upper = mean_diff + t_crit * se
else:
    se = ci_lower = ci_upper = np.nan


pd.options.display.float_format = "{:,.2f}".format
print("\nSummary of paired values (per country):")
print(paired[['pre','post','diff']])

print("\nSummary statistics:")
print(f"Median pre: {median_pre:.2f} kg; Median post: {median_post:.2f} kg; Median diff: {median_diff:.2f} kg")
print(f"Mean difference (post - pre): {mean_diff:.2f} kg; SE: {se:.2f}")
print(f"95% CI for mean difference: [{ci_lower:.2f}, {ci_upper:.2f}]")

print("\nWilcoxon signed-rank test (two-sided):")
print(f"  statistic = {w_stat if not np.isnan(w_stat) else 'NA'}, p-value = {w_p if not np.isnan(w_p) else 'NA'}")

print("\nPaired t-test (two-sided):")
print(f"  t = {t_stat if not np.isnan(t_stat) else 'NA'}, p-value = {t_p if not np.isnan(t_p) else 'NA'}")


#welp, neither the wilcoxon signed-rank test nor the paired t-test were significant. The sharpiro walk test also showed that the distirbution is not normal. 
```
